```python
#importing libraraies
import bt
import pandas as pd
import numpy as np
import re
import plotly as plt
from time import sleep
import random
import yfinance as yf
import csv
from sklearn.linear_model import Ridge
```

```python
prices = pd.read_csv("C:\\Users\\eitan\\Downloads\\data science in finance\\yfinance data 2007b by year done completly.csv")
#removing doubles
col_names = []
for column in prices:
    col_names.append(column)
black_list = []
def hasNumbers(col_names):
    for i in col_names:
        if any(char.isdigit() for char in i):
            black_list.append(i)
    return black_list
black_list = hasNumbers(col_names)
prices = prices.drop(black_list, axis=1)

prices['Date'] = pd.to_datetime(prices['Date'])
prices = prices.set_index(prices['Date'])
prices = prices.sort_index()
prices = prices['2007-1-1':'2011-12-30']
prices = prices.T
prices
```

```python
prices = prices.dropna(axis=0, how='all', thresh=None, subset=None, inplace=False)
prices = prices.iloc[1: , :]
prices 
```

```python
prices_t = prices.T
shift_prices = prices_t.pct_change().shift(1)
month = prices_t.groupby(pd.Grouper(freq="M")).nth(0)
month_change = month.pct_change().shift(1)
```
